```python
import pandas as pd

df = pd.read_csv("../data/a.us.txt")
df['Date'] = pd.to_datetime(df['Date'])
df.info()
```

```python
# Sample Visualizations on one of the Stock prices
import matplotlib.pyplot as plt

plt.plot(df['Date'], df['Close'])
```

```python
print(f"Max close price: {df['Close'].max():.2f}")
print(f"Min close price: {df['Close'].min():.2f}")
print(f"Avg close price: {df['Close'].mean():.2f}")
print(f"Standard Deviation close price: {df['Close'].std():.2f}")
```
